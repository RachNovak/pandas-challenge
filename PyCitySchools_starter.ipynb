# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
import csv

# File to Load (Remember to Change These)
school_data_to_load = os.path.join("Resources","schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
schooldata_df = pd.read_csv(school_data_to_load)
studentdata_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(schooldata_df, studentdata_df, how="left", on=["school_name"])
school_data_complete
```

## District Summary

```python
# Calculate the total number of unique schools
#create aary of school name
schoolnamescount = school_data_complete['school_name'].unique()
#find length and count
school_count = len(schoolnamescount)
school_count
```

```python
# Calculate the total number of students
dist_student_count = schooldata_df['size'].sum()
dist_student_count
#39170

#check data again student dataframe
#total_student_rec = school_data_complete['student_name'].count()
#total_student_rec
```

```python
# Calculate the total budget
total_budget = schooldata_df.loc[:,'budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_mathscore = studentdata_df['math_score'].mean()
average_mathscore
```

```python
# Calculate the average (mean) reading score
#average_readingscore = school_data_complete['reading_score'].mean()
#average_readingscore
#81.87778....

avg_readingscore = studentdata_df['reading_score'].mean()
avg_readingscore
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
#passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
#passing_math_percentage = passing_math_count / float(student_count) *100
#passing_math_percentage

#rewrite to a correctly working formual
passing_math_count = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()
perc_pass_math = passing_math_count/dist_student_count
perc_pass_math
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
perc_pass_reading = passing_reading_count /dist_student_count
perc_pass_reading
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /dist_student_count
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [dist_student_count],
    "Total Budget": [total_budget],
    "Average Reading Score": [avg_readingscore],
    "Average Math Score": [average_mathscore],
    "% Passing Reading":[perc_pass_reading],
    "% Passing Math": [perc_pass_math],
    "Overall Passing Rate": [overall_passing_rate]
}
)
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.1f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.1f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.1%}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.1%}".format)
district_summary["Overall Passing Rate"] = district_summary["Overall Passing Rate"].map("{:.1%}".format)


# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select the school type 
#group schools together
by_school = school_data_complete.set_index('school_name').groupby(['school_name'])
types = schooldata_df.set_index('school_name')['type']
types
```

```python
# Calculate the total student count
studentsperschool = by_school['Student ID'].count()
studentsperschool
```

```python
schoolbudget = schooldata_df.set_index('school_name')['budget']
schoolbudget
```

```python
# Calculate the total school budget and per capita spending
budgetperstudent = schooldata_df.set_index('school_name')['budget']/schooldata_df.set_index('school_name')['size']
budgetperstudent
```

```python
# Calculate the average test scores 
avgreading = by_school['reading_score'].mean()
avgmath = by_school['math_score'].mean()

avgreading
#avgmath
```

```python
# Calculate the number of schools with math scores of 70 or higher
#schoolspassingmath = avgmath[avgmath >= 70].count()
#schoolspassingmath

schoolpassingmathavg = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/studentsperschool
schoolpassingmathavg
```

```python
# Calculate the number of schools with reading scores of 70 or higher
schoolpassingreadingavg = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/studentsperschool
schoolpassingreadingavg
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passingboth = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/studentsperschool 
passingboth

```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = schoolpassingmathavg.groupby(["school_name"]).count() / studentsperschool * 100
per_school_passing_math

per_school_passing_reading = schoolpassingreadingavg.groupby(["school_name"]).count() / studentsperschool * 100
per_school_passing_reading

overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/studentsperschool 
overall
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School Type: ": types,
    "Number of Students: ": studentsperschool,
    "School Budget: ": schoolbudget,
    "Spend per Student: ": budgetperstudent,
    "Average Math Score: ": avgmath,
    "Average Reading Score: ": avgreading,
    "% Passing Math: ": schoolpassingmathavg,
    "% Passing Reading: ": schoolpassingreadingavg,
    "Overall Passing Rate: ": passingboth
})
per_school_summary

# Formatting
per_school_summary["Number of Students: "] = per_school_summary["Number of Students: "].map("{:,}".format)
per_school_summary["School Budget: "] = per_school_summary["School Budget: "].map("${:,.2f}".format)
per_school_summary["Spend per Student: "] = per_school_summary["Spend per Student: "].map("${:,.2f}".format)
per_school_summary["Average Math Score: "] = per_school_summary["Average Math Score: "].map("{:,.2f}".format)
per_school_summary["Average Reading Score: "] = per_school_summary["Average Reading Score: "].map("{:,.2f}".format)
per_school_summary["% Passing Math: "] = per_school_summary["% Passing Math: "].map("{:,.1%}".format)
per_school_summary["% Passing Reading: "] = per_school_summary["% Passing Reading: "].map("{:,.1%}".format)
per_school_summary["Overall Passing Rate: "] = per_school_summary["Overall Passing Rate: "].map("{:,.1%}".format)


# Display the DataFrame
per_school_summary
```
